```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/electric vhicle data set.csv')
```

## Univariate Analysis

Categorical Variables


State Distribution

```python
# Count of vehicles per state
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='State', order=df['State'].value_counts().index)
plt.title('Vehicle Distribution by State')
plt.xticks(rotation=90)
plt.show()
```

Brand Distribution

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Make', order=df['Make'].value_counts().index)
plt.title('Vehicle Distribution by Make')
plt.xticks(rotation=90)
plt.show()
```

Electric Vehicle Type Distribution

```python
vehicle_type_counts = df['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Electric Vehicle Types')
plt.show()
```

CAFV Eligibility

```python
# Bar plot for CAFV Eligibility
plt.figure(figsize=(6,6))
sns.countplot(data=df, x='Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.title('CAFV Eligibility')
plt.show()
```

Numerical Variables

Electric Range Distribution

```python
# Histogram of Electric Range
plt.figure(figsize=(8,6))
sns.histplot(df['Electric Range'], bins=30, kde=True)
plt.title('Electric Range Distribution')
plt.show()
```

Model Year Distribution

```python
# Histogram of Model Year
plt.figure(figsize=(8,6))
sns.histplot(df['Model Year'], bins=20, kde=True)
plt.title('Model Year Distribution')
plt.show()
```

## Bivariate Analysis

Categorical vs. Categorical


State vs. Electric Vehicle Type

```python
# Stacked bar plot: State vs Electric Vehicle Type
plt.figure(figsize=(10,6))
pd.crosstab(df['State'], df['Electric Vehicle Type']).plot(kind='bar', stacked=True, figsize=(12,7))
plt.title('Electric Vehicle Type Distribution by State')
plt.xticks(rotation=90)
plt.show()
```

CAFV Eligibility vs. Electric Vehicle Type

```python
# Bar plot for CAFV Eligibility by Electric Vehicle Type
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Electric Vehicle Type', hue='Clean Alternative Fuel Vehicle (CAFV) Eligibility')
plt.title('CAFV Eligibility by Electric Vehicle Type')
plt.show()
```

Numerical vs. Numerical


Electric Range vs. Model Year

```python
# Scatter plot of Electric Range vs Model Year
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Model Year', y='Electric Range')
plt.title('Electric Range vs Model Year')
plt.show()
```

Electric Range vs. Base MSRP

```python
# Scatter plot of Electric Range vs Base MSRP
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Base MSRP', y='Electric Range')
plt.title('Electric Range vs Base MSRP')
plt.show()
```

Categorical vs. Numerical

State vs. Electric Range

```python
# Box plot of Electric Range by State
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='State', y='Electric Range')
plt.title('Electric Range by State')
plt.xticks(rotation=90)
plt.show()
```

Electric Vehicle Type vs. Electric Range

```python
# Box plot of Electric Range by Electric Vehicle Type
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Electric Vehicle Type', y='Electric Range')
plt.title('Electric Range by Electric Vehicle Type')
plt.show()
```

## Choropleth using plotly.express

```python
import plotly.express as px

# Count the number of EVs by State (you can also group by County if needed)
ev_count_by_state = df['State'].value_counts().reset_index()
ev_count_by_state.columns = ['State', 'EV Count']

# Create the choropleth map
fig = px.choropleth(
    ev_count_by_state,
    locations='State',
    locationmode='USA-states',  # Use 'USA-states' for US states
    color='EV Count',
    color_continuous_scale='Viridis',
    scope='usa',
    labels={'EV Count': 'Number of EV Vehicles'},
    title='Number of Electric Vehicles by State'
)

# Show the choropleth map
fig.show()
```

## Racing BAr Plot

```python


# Filter relevant columns (if necessary)
# Create a new DataFrame with the count of EVs by make and year
ev_count = df.groupby(['Model Year', 'Make']).size().reset_index(name='Count')

# Create the racing bar plot
fig = px.bar(ev_count,
             x='Count',
             y='Make',
             color='Make',
             animation_frame='Model Year',
             range_x=[0, ev_count['Count'].max() + 10],  # Adjust range for better visibility
             title='Racing Bar Plot of Electric Vehicles by Make Each Year',
             labels={'Count': 'Number of EVs', 'Make': 'Make of EV'},
             orientation='h')  # Horizontal bar plot

# Show the animated bar plot
fig.show()
```

```python

# Data preparation: Count the number of EVs by Make and Model Year
ev_count = df.groupby(['Model Year', 'Make']).size().reset_index(name='Count')

# Create a racing bar plot with improved customization
fig = px.bar(
    ev_count,
    x='Count',
    y='Make',
    color='Make',
    animation_frame='Model Year',
    animation_group='Make',  # Ensure Make stays consistent across frames
    range_x=[0, ev_count['Count'].max() + 50],  # Extend range for readability
    title='Electric Vehicles Count by Make Over the Years',
    labels={'Count': 'Number of EVs', 'Make': 'Make of EV'},  # Custom axis labels
    orientation='h',  # Horizontal bar plot
    color_discrete_sequence=px.colors.qualitative.Vivid  # Improved color palette
)

# Update layout for better aesthetics
fig.update_layout(
    xaxis_title='Number of Electric Vehicles',
    yaxis_title='Vehicle Make',
    title_font_size=24,
    title_font_color='darkblue',
    plot_bgcolor='whitesmoke',  # Light background
    paper_bgcolor='lightgray',  # Page background
    font=dict(size=14),
    showlegend=False  # Hide the legend for simplicity
)

# Add animation speed and smoother transitions
fig.update_traces(texttemplate='%{x}', textposition='outside')  # Show count on bars
fig.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'showactive': False,
        'buttons': [{
            'label': 'Play',
            'method': 'animate',
            'args': [None, {
                'frame': {'duration': 800, 'redraw': True},
                'fromcurrent': True,
                'transition': {'duration': 500, 'easing': 'linear'}
            }]
        }]
    }]
)

# Show the final racing bar plot
fig.show()
```
